Break, Continue, Else, Pass

```python
#Break - terminate execution of loop

counter = 0
while counter <= 10:
    print(counter)
    if counter == 6:
        break
    counter += 1
```

```python
integers = [1,2,3,4,5]

for num in integers:
    power_of_num = num**num
    if power_of_num > 3000:
        break
    print(power_of_num)
```

```python
#Continue - to skip the rest of the code inside the current loop iteration and move on to the next iteration.

number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
```

```python
integers = [1,2,3,4,5,6,7]

for num in integers:
    power_of_num = num**num
    if 200 <= power_of_num <= 3250:
        continue
    print(num, power_of_num)
```

```python
#Else - can be used with a while loop and it is executed when the condition becomes false.

counter = 0
while counter <= 10:
    if counter == 6.5:
        break
    print(counter)
    counter += 1
else:
    print('No break found')
```

```python
integers = [1,2,3,4,5]

for num in integers:
    power_of_num = num**num
    if power_of_num > 4000:
        break
    print(power_of_num)
else:
    power_of_num = num**num
    print(power_of_num)
```

```python
#Pass - placeholder

integers = [1,2,3,4,5]

for num in integers:
    pass
```

Nested While Loops

```python
number = 0

while number <= 10:
    print(number)
    inner_number = 0
    while inner_number < 5:
        print(inner_number)
        inner_number += 1
    number += 1
```

```python
number = 0

while number <= 10:
    print(number)
    inner_number = 0
    while inner_number < 5:
        print(inner_number)
        inner_number += 1
    else:
        print('Inner While Loop Ended')
    number += 1
else:
    print('Outer While Loop Ended')
```

```python
number = 0

while number <= 10:
    print(number)
    inner_number = 0
    while inner_number < 5:
        print(inner_number)
        if inner_number == 2:
            break
        inner_number += 1
    else:
        print('Inner While Loop Ended')
    number += 1
else:
    print('Outer While Loop Ended')
```

```python
number = 0

while number <= 10:
    print(number)
    inner_number = 0
    while inner_number < 5:
        inner_number += 1
        if inner_number == 2:
            continue
        print(inner_number)
    else:
        print('Inner While Loop Ended')
    number += 1
else:
    print('Outer While Loop Ended')
```

List Comprehension

```python
flavors = ['Vanilla', 'Choco', 'Choco Fudge', 'Strawberry', 'Choco Swirl']
choco_flavors = []

for flavor in flavors:
    if 'Choco' in flavor:
        choco_flavors.append(flavor)
print(choco_flavors)
```

```python
choco_flavor2 = [flavor for flavor in flavors if 'Choco' in flavor]
print(choco_flavor2)
```

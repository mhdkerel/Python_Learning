```python
import re
```

```python
quote = 'My name is Mr. Freberg. My phone number is 542-234-6548. My email is AlexFreberg@gmaiil.com'
```

```python
re.findall("\d", quote)
```

```python
# \w - Word Character

re.findall("\w", quote)
```

```python
# \W - Not a Word Character

re.findall("\W", quote)
```

```python
# \A - Start of a String, same with ^

re.findall("\A[A-Z]", quote)
```

```python
# \Z - End of a String, same with $

re.findall("[a-zA-Z]\Z", quote)
```

```python
# \b - Word boundry

quote = 'I like Python is PythonLove a word?'

re.findall(r"\bPython", quote)
```

```python
# \B - NOT a Word boundry

quote = 'I like Python is PythonLove a word?'

re.findall(r"\Bython\B", quote)
```

```python
string = 'There are 365 days in a year. 8760 hours. 525600 minutes. Only 24 days until Christmas. 1 Day at a time'

re.findall(r"\d+", string)
```

```python
len(re.findall(r"\d+", string))
```

Regex Modifiers

```python
import re
```

```python
quote = 'My name is Mr. Freberg. My phone number is 542-234-6548. My email is AlexFreberg@gmail.com'
```

```python
re.findall(r"\b[a-z]+\b",quote)
```

```python
re.findall(r"\b[a-z]+\b",quote, re.I)
```

```python
target_string = 'Alex likes the number 16\nChristine likes the number 24'

target_string2 = """Alex likes the number 16
Christine likes the number 24
"""
```

```python
re.findall(r"^\w{4}", target_string2)
```

```python
re.findall(r"^\w{4}", target_string2, re.M)
```

```python
re.findall(r"\d{2}$", target_string2, re.M)
```

```python
string = 'I like ice cream.\nI like pie as well!'
```

```python
re.findall(r".+", string)
```

```python
re.findall(r".+", string, re.S)
```

```python
result = re.search(r".+", string)

print(result.group())
```

```python
result = re.search(r".+", string, re.S)

print(result.group())
```

```python
string = 'Alex has how many dogs? 75'
```

```python
result = re.search(r"""(^\w{4}) #This needs to be a 4 letter word at the beginning
.+(\d{2}$) # This needs to be a 2 digit number at the end
""", string, re.X)
```

```python
result.group(1)
```

```python
result.group(2)
```

Regex Use Cases

```python
import re

random_text = '''
My name is Mr. Freberg. My phone number is 542-234-2819. My email is AlexFreberg@gmail.com
My name is Mr. Max. My phone number is 413-234-2568. My email is RoughRuffy@yahoo.com.
My name is Mrs. Rosie. My phone number is 285-036-8215. My email is GoldenGirl@apple.com.
'''
```

```python
re.findall(r"@([a-zA-Z]+)",random_text)
```

```python
re.findall(r"@([\w\.+]+)",random_text)
```

```python
re.findall(r"[\w\.+]+@[\w\.+]+",random_text)
```

```python
random_text = '''
My name is Mr. Freberg. My phone number is 542-234-2819. My email is AlexFreberg@gmail.com
My name is Mr. Max. My phone number is 413-234-2568. My email is RoughRuffy@yahoo.com.
My name is Mrs. Rosie. My phone number is 285-036-8215. My email is GoldenGirl@apple.com.
'''
```

```python
re.findall(r"\d{3}-\d{3}-\d{4}",random_text)
```

```python
re.findall(r"\ (\d{3})",random_text)
```

```python
email_list = ['AlexFreberg@gmail.com', 'RoughRuffy@yahoo.com', 'GoldenGirl@apple.com']
```

```python
for email in email_list:
    result = re.findall(r"@([a-zA-Z]+)",email)
    print(result)
```

```python
domain_list = [re.findall(r"@([a-zA-Z]+)",email)[0] for email in email_list]

print(domain_list)
```

```python
import os
import pandas as pd
from bs4 import BeautifulSoup

# 1. Configuration
input_folder = r"C:\Users\khairil.akub\Desktop\Mohd Khairil\Project\Python\CAR Data Extraction\New Code\HTML Files"
output_folder = r"C:\Users\khairil.akub\Desktop\Mohd Khairil\Project\Python\CAR Data Extraction\New Code\CAR_Extraction_Output"
os.makedirs(output_folder, exist_ok=True)

# 2. Helper function - Extract key-value data from table
def extract_key_value_pairs(table):
    data = {}
    rows = table.find_all("tr")

    for i, row in enumerate(rows):
        cells = [td.get_text(strip=True) for td in row.find_all("td")]
        if not cells:
            continue


        # case 1: side by side
        if len(cells) == 2:
            key, val = cells
            if key and val:
                data[key] = val


        # case 2: top-below format
        elif len(cells) == 1:
            key = cells[0]
            next_row = row.find_next_sibling("tr")
            if next_row:
                next_cells = [td.get_text(strip=True) for td in next_row.find_all("td")]
                if len(next_cells) == 1:
                    val = next_cells[0]
                    if key and val:
                        data[key] = val

    return data

# 3. Main extraction loop for all HTML files
master_records = []
section_dfs = {}

for file_name in os.listdir(input_folder):
    if not file_name.lower().endswith(".html"):
        continue

    file_path = os.path.join(input_folder, file_name)
    print(f"Processing {file_name}")

    with open(file_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    #sections = soup.find_all(["h1", "h2", "h3"])
    all_tables = soup.find_all("table")
    car_no = None

    #for section in sections:
        #section_title = section.get_text(strip=True)
        #table = section.find_next("table")
       # if not table:
           # continue

    for idx, table in enumerate(all_tables, start=1):
        heading_tag = table.find_previous(["h1", "h2", "h3"])
        section_title = heading_tag.get_text(strip=True) if heading_tag else f"Section_{idx}"
    
    
    section_data = extract_key_value_pairs(table)

    #capture CAR No once if present
    if "CAR No" in section_data and not car_no:
        car_no = section_data["CAR No"]

    section_data["CAR No"] = car_no
    section_data["Section"] = section_title
    section_data["FileName"] = file_name

    master_records.append(section_data)

    #store section-wise DataFrame for individuals CSVs
    if section_title not in section_dfs:
        section_dfs[section_title] = []
    section_dfs[section_title].append(section_data)

# 4. Combine all into a Master CSV
if master_records:
    master_df = pd.DataFrame(master_records)
    master_csv = os.path.join(output_folder, "CAR_Master_Extract.csv")
    master_df.to_csv(master_csv, index=False, encoding='utf-8-sig')
    print(f"Master file saved: {master_csv}")

#5. Save individual csvs per section
for section_title, records in section_dfs.items():
    clean_name = section_title.replace(" ", "_").replace("/", "_")
    section_df = pd.DataFrame(records)
    section_csv = os.path.join(output_folder, f"{clean_name}.csv")
    section_df.to_csv(section_csv, index=False, encoding="utf-8-sig")
    print(f"Section file saved: {section_csv}")

print("\nExtraction complete for all HTML files!")
```

```python
phone_numbers = ['2332423412','212312312312312','1231231231123123']

for num in phone_numbers:
    if len(num) == 10:
        print(num)
```

```python
phone_numbers = [2332423412,212312312312312,1231231231123123]

for num in phone_numbers:
    if len(str(num)) == 10:
        print(num)
```

Format

```python
name = 'Alexander'

ID = '12323523'

print('My name is ' + name + '. My ID is ' + ID)
```

```python
print("My name is {}. My ID is {}".format(name, ID))
```

```python
first_name = 'Mohd Khairil'
last_name = 'Akub'

print("My name is {1} {0}.".format(first_name, last_name))
```

```python
name = 'Alex'
adjective = 'weird'

print("{name} is a {genius} guy.".format(name = name, genius = adjective))
```

```python
height = 177.12231

print("I am {:.3f} centimeters tall.".format(height))
```

```python
my_list = ['0Rocky Road','1Chocolate','2Strawberry']

new_list = [word.strip('0123456789') for word in my_list]

print(new_list)
```
